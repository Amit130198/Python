```python
# Question 1 and 2
Name = input("Enter your name: ")
Age = int(input("Enter you age: "))
subject_A = int(input("Enter you marks: "))
subject_B = int(input("Enter you marks: "))

Obtained_Marks = subject_A + subject_B
print(f"My Name is {Name} and I'm {Age}years old. I'm studying MBA in from Pune University wherein I have scored {Obtained_Marks}  ")
```

```python
#Question 3
Product_name = input("Enter Product Name: ")
Quantity = int(input("Select the Quantity: "))
Price = int(input("Enter the Amount: "))
Grand_Total = Quantity * Price

print(f"Your Final Amount of {Product_name} is {Grand_Total}")
```

```python
#Question58
Student_name = input("Enter Student Name: ")
English_Marks = int(input("Enter Subject Marks: "))
Maths_Marks = int(input("Enter Subject Marks: "))
Science_Marks = int(input("Enter Subject Marks: "))
Total_Marks = English_Marks + Maths_Marks + Science_Marks
Average_Marks = (Total_Marks / 150) * 100

print(f"Average marks of {Student_name} is {Average_Marks:.2f}")
```
